```python
import numpy as np
import pandas as pd

threshold = 0.01
succ_adj = 10

data = []

for subject, project, gt in prog_list:
    os.chdir(os.path.join(base_dir, f"{project}-modified"))
    num_sample_list = []
    esti_list = []
    time_list = []
    for i in range(multiple_run):
        result_path = f"{subject}-result-{threshold}-{i}"
        print(result_path, gt, end="\r")
        with open(result_path, "r") as f:
            lines = f.readlines()
            time_spent = lines[-3].strip().split("\t")[-1]
            m, s = time_spent[:-1].split("m")
            m, s = int(m), float(s)
            time_list.append(m * 60 + s)
            if gt > 0:
                num_sample_needed = int(lines[-5].split()[0])
                num_sample_list.append(num_sample_needed)
                esti = float(lines[-5].split()[3])
                esti_list.append(esti)
    if gt > 0:
        num_sample = (np.mean(num_sample_list), np.std(num_sample_list))
        esti = (np.mean(esti_list), np.std(esti_list))
    else:
        num_sample = (2000000, 0)
        esti = (10 ** -6, 0)
    time_spent = (np.mean(time_list), np.std(time_list))

    data.append([project, subject, gt, num_sample[0], num_sample[1], esti[0], esti[1], time_spent[0], time_spent[1]])

df = pd.DataFrame(data, columns=["project", "subject", "gt", "num_sample", "num_sample_std", "esti", "esti_std", "time_spent", "time_spent_std"])
display(df)
```
