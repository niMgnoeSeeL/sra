```python
import pandas as pd

import os
os.chdir("") # set the root directory of the project

crash_data_path = "fuzz-data/crashes.csv"
crash_data = pd.read_csv(crash_data_path)
```

```python
jasper_crash = crash_data[crash_data["Subject"] == "jasper"]
max_trace_length = max(jasper_crash["Trace_Length"])
max_length_crash = jasper_crash[
    jasper_crash["Trace_Length"] == max_trace_length
].drop_duplicates(subset=["Function", "Line"])
display(max_length_crash)
max_length_crash = jasper_crash[
    jasper_crash["Trace_Length"] == max_trace_length - 1
].drop_duplicates(subset=["Function", "Line"])
display(max_length_crash)
max_length_crash = jasper_crash[
    jasper_crash["Trace_Length"] == max_trace_length - 2
].drop_duplicates(subset=["Function", "Line"])
display(max_length_crash)
max_length_crash = jasper_crash[
    jasper_crash["Trace_Length"] == max_trace_length - 3
].drop_duplicates(subset=["Function", "Line"])
display(max_length_crash)
max_length_crash = jasper_crash[
    jasper_crash["Trace_Length"] == max_trace_length - 4
].drop_duplicates(subset=["Function", "Line"])
display(max_length_crash)
```
